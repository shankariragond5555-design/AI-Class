Trim Whitespace Methods

1. lstrip(): It removes leading/left whitespaces

```python
text1= "     Hello, world     "
```

```python
print(text1) # Output : "Hello, world"
```

```python
result=text1.lstrip()
```

```python
print(result)
```

2. rstrip(): It removes right hand side whitespaces

```python
text2= "    Hello, world    "
```

```python
print(text2)
```

```python
result1=text2.rstrip()
```

```python
print(result1)
```


3. strip(): It removes leading & trailing (left and right hand side) whitespaces

```python
text3= "    Hello, world    "
```

```python
print(text3)
```

```python
result2=text3.strip()
```

```python
print(result2)
```

Split Methods

1. split(): It divides string based on delimiter

```python
str1="hello, welcome to india"
```

```python
words=str1.split(" ")
```

```python
print(words)
```

```python
str2="hello, welcome to india"
```

```python
words1=str2.split(",")
```

```python
print(words1)
```

2.partition() : It divides string into three parts, 1. before separator 2. separator 3. after separator

```python
str3="who are you?"
```

```python
result3=str3.partition("are")
```

```python
print(result3)
```

2. upper() - converts to uppercase

```python
str5="hello"
```

```python
result6=str5.upper()
```

```python
print(result6)
```

3. capitalize() : It converts first character to uppercase , rest lowercase

```python
str6="hello world"
```

```python
result7=str6.capitalize()
```

```python
print(result7)
```

4. title() : converts first char. of each word to uppercase

```python
str7="hello world"
```

```python
result8=str7.title()
```

```python
print(result8)
```

swapcase() : swaps uppercase <-> lowecase

```python
str8="hello WORLD"
```

```python
result9=str8.swapcase()
```

```python
print(result9)
```
