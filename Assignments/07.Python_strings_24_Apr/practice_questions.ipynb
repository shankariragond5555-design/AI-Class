```python
#Q2. What will be the output of both print statements? What does lstrip() remove?

raw_name = "   Shankar Iragond   "
print(raw_name)      
stripped_left = raw_name.lstrip()  #lstrip() remove whitespaces from left side of the string
print(stripped_left)  # Output :" Shankar Iragond"

#Explanation: The lstrip() method removes any leading characters (space is the default leading character to remove) from the left side of the string. In this case, it removes the leading spaces before "Shankar Iragond". 
```

```python
#Q3. Predict the output. What does rstrip() keep and what does it remove?

padded_city = "   Satara   "
stripped_right = padded_city.rstrip() #rstrip() keeps whitespaces from left side and remove it from right side
print(stripped_right) # Output :" Satara"

#Explanation: The rstrip() method removes any trailing characters (space is the default trailing character to remove) from the right side of the string. In this case, it removes the trailing spaces after "Satara".   
```

```python
#Q4. What will be the output? Explain the difference between strip() and lstrip().

user_input = "   Data Science   "
clean_input = user_input.strip() #strip() removes whitespaces from both side and lstrip() removes whitespaces from left side
print(clean_input) # Output :"Data Science"

#Explanation: The strip() method removes any leading and trailing characters (space is the default character to remove) from both sides of the string. In this case, it removes the leading and trailing spaces around "Data Science". On the other hand, lstrip() would only remove the leading spaces, leaving the trailing spaces intact.    
```

```python

#Q5. Predict the output. What data type does split() return?

address = "Mumbai Pune Delhi Chennai"
city_list = address.split(" ") # split returns in list data type
print(city_list) # Output : ['Mumbai', 'Pune', 'Delhi', 'Chennai']
print(type(city_list)) # Output : <class 'list'>

#Explanation: The split() method takes a string and splits it into a list of substrings based on the specified delimiter (in this case, a space " "). The resulting data type is a list, which contains the individual city names as its elements.  
```

```python
#Q6. What will be the output? How many parts does partition() always return?

bio = "My name is Shankar"
bio_parts = bio.partition("is") 
print(bio_parts) # Output : "My name","is","Shankar"
#Explanation: The partition() method splits the string into three parts based on the specified separator (in this case, "is"). It returns a tuple containing the part before the separator, the separator itself, and the part after the separator. In this example, it splits "My name is Shankar" into "My name", "is", and "Shankar".   
```

```python

#Q7. What will be the output of both print statements?

department = "COMPUTER SCIENCE"
dept_lower = department.lower()
print(dept_lower)  # Output : computer science

subject = "machine learning"
subject_upper = subject.upper()
print(subject_upper)  # Output : MACHINE LEARNING

#Explanation: The lower() method converts all characters in the string to lowercase, while the upper() method converts all characters to uppercase. In this case, "COMPUTER SCIENCE" is converted to "computer science" and "machine learning" is converted to "MACHINE LEARNING".
```

```python
#Q8. Predict the output. What exactly does capitalize() do to the rest of the string?

college_name = "Shivaji university kolhapur"
formatted_college = college_name.capitalize()
print(formatted_college) #Output : "Shivaji university kolhapur"

#Explanation: The capitalize() method converts the first character of the string to uppercase and leaves the rest of the characters unchanged. In this case, only the first letter "S" in "Shivaji" is capitalized, while the rest of the string remains in lowercase.
```

```python
#Q9. What will be the output? How is title() different from capitalize()?

book_title = "the art of python programming"
proper_title = book_title.title() #title() convert the first letter of every word into uppercase and capitalize() convert the first letter in uppercase of first word only
print(proper_title) # Ouput : "The Art Of Python Programming"

#Explanation: The title() method converts the first character of each word in the string to uppercase and the rest of the characters to lowercase. In this case, it converts "the art of python programming" to "The Art Of Python Programming". On the other hand, capitalize() would only convert the first character of the entire string to uppercase, resulting in "The art of python programming".    
```

```python
#Q10. Predict the output. What does swapcase() do to each character?

mixed_text = "pYTHON iS fUN"
swapped = mixed_text.swapcase()
print(swapped) #Output : "Python IS Fun"

#Explanation: The swapcase() method converts all uppercase characters in the string to lowercase and all lowercase characters to uppercase. In this case, "pYTHON iS fUN" is transformed to "Python IS Fun", where each character's case is swapped.
```

```python
#Q11. What will be the output of all three print statements?

marks = 95
percentage = 87.5
comp_num = 3+2j

marks_str = str(marks)
percentage_str = str(percentage)
comp_str = str(comp_num)

print(type(marks_str))   # all print statments are converted into string so Output : <class 'str'>
print(type(percentage_str))
print(type(comp_str))

#Explanation: The str() function converts the given value into a string data type. In this case, the integer 95, the float 87.5, and the complex number 3+2j are all converted to their string representations. Therefore, the output of all three print statements will indicate that the resulting data type is <class 'str'>.
```

```python
#Q12. What will be the output? What does chr() take as input?

letter1 = chr(66) # chr() takes unicode value as a input
print(letter1) # Output : B

letter2 = chr(98)
print(letter2) # Output : b

#Explanation: The chr() function takes an integer representing a Unicode code point and returns the corresponding character as a string. In this case, chr(66) returns 'B' because 66 is the Unicode code point for the uppercase letter 'B', and chr(98) returns 'b' because 98 is the Unicode code point for the lowercase letter 'b'.
```

```python
#Q13. What will be the output? What does ord() take as input?

code1 = ord("Z") # ord() takes characters as a input 
print(code1) #Output : 90

code2 = ord("z")
print(code2) #Output : 122

#Explanation: The ord() function takes a single character as input and returns its Unicode code point as an integer. In this case, ord("Z") returns 90 because 'Z' is the Unicode code point for the uppercase letter 'Z', and ord("z") returns 122 because 'z' is the Unicode code point for the lowercase letter 'z'.
```

```python
#Q14. Predict the output. What does join() use as the separator here?

skills = ["Python", "SQL", "Machine Learning", "Excel"]
skill_set = ", ".join(skills) # join() used , as a separator here
print(skill_set) #Output : Python, SQL, Machine Learning, Excel

#Explanation: The join() method takes an iterable (in this case, a list of skills) and concatenates its elements into a single string, using the specified separator (in this case, ", ") between each element. Therefore, the output will be a string that lists all the skills separated by a comma and a space.
```

```python
#Q15. What will be the output of all three prints?

employee_data = "emp:Suresh:dept:HR"
updated_data = employee_data.replace("Suresh", "Ganesh")
print(updated_data) # Output : "emp:Ganesh:dept:HR"

split_data = employee_data.split(":")
print(split_data) # Output : 'emp', 'Suresh', 'dept', 'HR'

partition_data = employee_data.partition(":")
print(partition_data) # Output : 'emp', ':', 'Suresh:dept:HR'

#Explanation: The replace() method replaces occurrences of a specified substring with another substring. In this case, it replaces "Suresh" with "Ganesh". The split() method splits the string into a list of substrings based on the specified delimiter (":"). The partition() method splits the string into three parts based on the first occurrence of the specified separator (":") and returns a tuple containing the part before the separator, the separator itself, and the part after the separator.
```
