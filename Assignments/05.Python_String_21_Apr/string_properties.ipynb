2. Concentation

```python
greeting="Hello"
name="Shankar"
message=greeting+name
```

```python
print(message)
```

```python
print(type(message))
```

```python
space=" "
```

```python
message1=greeting+space+name
```

```python
print(message1)
```

```python
message2="Hello"+" "+"Shankar"
```

```python
print(message2)
```

3. Repetition

```python
word="Hello "
repeated_word=word*3
print(repeated_word)
```

```python
word="Hello"
word1=word+" "
repeated_word=word1*3
print(repeated_word)
```

4. Membership Testing (in/ not in)

```python
text="Python Programming"
print("Python" in text)
```

```python
print("Java" in text)
```

```python
print("Python" not in text)
```

```python
print("Java" not in text)
```
