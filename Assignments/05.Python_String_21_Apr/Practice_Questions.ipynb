```python
# 1. Explain why strings are immutable in Python.

# Strings are immutable because we cant change or modify the string once it created , when we try to modify the same string again it will change the memory address

# for ex;

text="Shankar"
print("Memory address before modification:",id(text))

# after modify it 

text="Shankar Iragond"
print("Memory address after modification :",id(text))
```
