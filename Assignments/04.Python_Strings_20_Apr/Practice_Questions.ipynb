```python

#5.	str1 = "Python" — Print the first character, last character, and middle character using positive indexing.
str1 = "Python"
print(str1[0]+str1[3]+str1[5])  # First character
```

```python
#6.	str1 = "Python" — Print the same characters using negative indexing.
str1 = "Python"
print(str1[-6]+str1[-3]+str1[-1])  # First character using negative indexing
```

```python
#7.	str2 = "welcome to india" — What is str2[4]? What is str2[-4]? Write the code and verify.
str2 = "welcome to india"
print(str2[4])   # Output: 'o'
print(str2[-4])  # Output: 'd'
```

```python
#8.	str3 = "Hello World" — Print the character at index 6 using both positive and negative index.
str3 = "Hello World"
print(str3[6])   # Output: 'o'
print(str3[-5])  # Output: 'o'      
```

```python
#9.	Given str4 = "DataScience", find which index gives you the letter S (both positive and negative).
str4 = "DataScience"
print(str4[4]) 
print(str4[-6])  # Output: 'S'  
```

```python
#10. str5 = "Rajesh" — Print each character one by one using its positive index (0 to 5).
str5 = "Rajesh"
print(str5[0])  # Output: 'R'
print(str5[1])  # Output: 'a'       
print(str5[2])  # Output: 'j'
print(str5[3])  # Output: 'e'
print(str5[4])  # Output: 's'
print(str5[5])  # Output: 'h'
```

```python
Slicing
```

```python
#11.	str6 = "welcome to india" — Slice and print only "welcome".

str6 = "welcome to india"
print(str6[0:7])
```

```python
#12.	From the same string, slice and print only "india".

str6 = "welcome to india"
print(str6[11:16])
```

```python
#13.	From str6 = "welcome to india", use a step of 2 to print every alternate character from index 0 to 7.

str6 = "welcome to india"
print(str6[0:7:2])
```

```python
#14.	str7 = "Artificial Intelligence" — Slice "Intelligence" from this string.

str7 = "Artificial Intelligence"
print(str7[11:23])
```

```python
#15.	From str7 = "Artificial Intelligence", print the entire string using [::], then again using [0:23:1]. Confirm the outputs match.

str7 = "Artificial Intelligence"

print(str7[::])
print(str7[0:23:1])
```
