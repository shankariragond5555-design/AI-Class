```python
from turtle import*
from colorsys import*
tracer(50)
bgcolor("black")
pensize(2)
h= 0
for i in range(350):
    h+= 0.0021
    c= hsv_to_rgb(1,h,1)
    color(c)
    goto(0,0)
    circle(180-i/4,90)
    circle(40)
    circle(180-i/4-60)
    right(50)
    pencolor(c)
    forward(i*3)
    left(59)
done()
```
