# Looping in tuple

# 1. while loop

```python
my_tuple=(10,20,30)
print(len(my_tuple))
```

```python
my_tuple=(10,20,30)
index=0

while (index<len(my_tuple)):
    print(my_tuple[index])
    index=index+1
```

```python
my_tuple=(10,20,30,10,20,50)
index=0
count=0

while (index<len(my_tuple)):
    print(my_tuple[index])
    index=index+1
    count=count+1
    print(count)
```

# 2. for loop

```python
my_tuple=(10,20,30)
```

```python
for i in my_tuple:
    print(i)
```

# 3.enumerate()

```python
my_tuple=(10,20,30)
```

```python
for i,j in enumerate(my_tuple):
    print("value is:",j,"and index is:",i)
```
