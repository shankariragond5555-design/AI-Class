## Function syntax

```python
print("program started")

def summation():
    num1=10
    num2=20
    result=num1+num2
    print(result)

summation()

print("program execution completed")
```

```python
def greet():
    print("Hello")
    print("program execution completed")
greet()
greet()
greet()
```

## Parameters and arguments

```python
print("program started")

def cal_sum(x,y,z):
    result=x+y+z
    print(result)

final_result=cal_sum(10,20,30)
print(final_result)
print("program ends")
    
```

## values from the user

```python
print("program started")

def cal_sum(x,y,z):
    result=x+y+z
    print(result)

num1=int(input("Enter first No:"))
num2=int(input("Enter second No:"))
num3=int(input("Enter third No:"))
cal_sum(num1,num2,num3)

print("program ends")
    
```
