```python
# Q10. What will be the output if user enters 50 and 30? What does return allow us to do with the result?
# print("program started")

# def find_max(num1, num2):
#     if num1 > num2:
#         return num1
#     else:
#         return num2

# a = int(input("Enter first number: "))
# b = int(input("Enter second number: "))
# biggest = find_max(a, b)
# print("Biggest number:", biggest)

# print("program ends")

print("program started")
def find_max(num1, num2):           
    if num1 > num2:        
        return num1    
    else:        
        return num2         
a = int(input("Enter first number: "))      
b = int(input("Enter second number: "))             
biggest = find_max(a, b)                
print("Biggest number:", biggest)
print("program ends")   
```
