```python
#Q12. Predict the output. What is the difference between isalpha() and isalnum()?

val1 = "Rupee100"
val2 = "Rupee"

print(val1.isalpha()) # it returns True if the string contains only alphabets
print(val1.isalnum()) # it returns True if the string contains both alphabets and digits
print(val2.isalpha())
print(val2.isalnum())

#Explanation
#isalpha() returns True if all characters in the string are alphabets (A-Z or   a-z) and there is at least one character. It returns False if there are any digits, spaces, or special characters.
#isalnum() returns True if all characters in the string are either alphabets or digits (A-Z, a-z, 0-9) and there is at least one character. It returns False if there are any spaces or special characters. 
```

```python
#Q13. What index does find() return for 'india' in the string below? Verify using indexing.

phrase = "welcome to india"
idx = phrase.find("india")
print(idx)
print(phrase[idx:idx+5])

#Explanation
#The find() method returns the lowest index of the substring if it is found in the given string. In this case, 'india' starts at index 11 in the string "welcome to india". The slicing phrase[idx:idx+5] retrieves the substring starting from index 11 to index 15, which is 'india'.    
```

```python
# Q14. What will be the output? Explain the role of len() as a built-in function vs a method.

product = "Computer"
print(len(product)) #len() is a built in function it gives us length of the string
print(max(product))
print(min(product))

#built in function is work on many data type but methods of the strings are specific they only work on string 
#len() is a built-in function that can be used with various data types, including strings, lists, tuples, etc., to return the number of items or characters. In contrast, methods like max() and min() are specific to certain data types (like strings) and perform operations based on the characteristics of that data type. 
```
