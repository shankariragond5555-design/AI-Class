## Aliasing - shallow copy

```python
list9=[10,20,30]
```

```python
print(list9)
```

```python
print(id(list9))
```

```python
list10=list9
```

```python
print(list10)
```

```python
print(id(list10))
```

```python
list9[0]=15
```

```python
print(list9)
```

```python
print(list10)
```

## Cloning - Deep Copy

```python
list9=[10,20,30]
```

```python
print(list9)
```

```python
list10=list9.copy()
```

```python
print(list10)
```

```python
print(id(list9))
print(id(list10))
```

```python
list9[0]=15
```

```python
print(list9)
```

```python
print(list10)
```
