```python
# Q1. What will be the output?
# num = 0
# while num < 5:
#     print("Number:", num)
#     num = num + 1
# print("Loop finished")

# Output:
num = 0
while num < 5:
    print("Number:", num)
    num = num + 1   
print("Loop finished")
```

```python
# Q2. What will be the output? What is the difference between < and <= in while loop?
# counter = 0
# while counter <= 5:
#     print("Counter:", counter)
#     counter = counter + 1
# print("Done")

counter = 0
while counter <= 5:
    print("Counter:", counter)
    counter = counter + 1
print("Done")
```

```python
# Q3. What will be the output?
# ticket = 1
# while ticket <= 3:
#     print("Ticket number:", ticket)
#     ticket = ticket + 1
# print("All tickets printed")

ticket = 1
while ticket <= 3:  
    print("Ticket number:", ticket)
    ticket = ticket + 1 
print("All tickets printed")
```

```python
# Q4. What will be the output? What does for loop do with a string?
# word = "Python"
# for letter in word:
#     print(letter)
# print("Loop completed")
word = "Python"
for letter in word: 
    print(letter)   
print("Loop completed") 
```

```python
# Q5. What will be the output?
# city = "Mumbai"
# for ch in city:
#     print(ch)
# print("Done")

city = "Mumbai"
for ch in city:     
    print(ch)   
print("Done")
```
