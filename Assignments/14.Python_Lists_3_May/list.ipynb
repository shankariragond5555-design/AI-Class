## while loop

```python
list3=[10,20,30,40,50]
print(len(list3))
```

```python
print("program stats")
list3=[10,20,30,40,50]

index=0
while (index<len(list3)):
    print(list3[index])
    index=index+1

print("program ends")
```

## For loop

```python
list4=[10,20,30,40,50]
```

```python
print("program starts")
for i in list4:
    print(i)

print("program ends")
```

## enumerate() function

```python
list5=[10,20,30,40,50]
```

```python
print("program starts")

for index,value in enumerate(list5):
    print(index,value)

print("program ends")
```

```python
print("program starts")

for index,value in enumerate(list5):
    print("Index is:",index,"and its value is:",value)

print("program ends")
```
