```python
# Q7. What will be the output? What does slicing return?
# numbers = [10, 20, 30, 40, 50]
# print(numbers[0:3])
# print(numbers[1:4])
# print(numbers[2:5])

numbers = [10, 20, 30, 40, 50]
print(numbers[0:3])     
print(numbers[1:4])     
print(numbers[2:5])         
        
```

```python
# Q8. What will be the output? What happens when start or end is omitted in slicing?
# prices = [100, 200, 300, 400, 500]
# print(prices[:3])
# print(prices[2:])
# print(prices[:])
# print(prices[0:5])

prices = [100, 200, 300, 400, 500]          
print(prices[:3])           
print(prices[2:])
print(prices[:])    
print(prices[0:5])
```

```python
# Q9. What will be the output? What does the step value do in slicing?
# marks = [10, 20, 30, 40, 50, 60, 70, 80]
# print(marks[0:8:2])
# print(marks[1:8:2])
# print(marks[::3])

marks = [10, 20, 30, 40, 50, 60, 70, 80]
print(marks[0:8:2])     
print(marks[1:8:2])     
print(marks[::3])
    
```
