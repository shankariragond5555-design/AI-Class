## Normal Method

```python
name="rahul"
age=25
print("Name:",name,"Age:",age)
# Name:rahul, Age:25
```

```python
print("My Name is :",name,",", "My Age is:",age)
# My Name is :rahul, My Age is:25
```

### 1. string Concatenation

```python
roll_no=100
address="pune"
roll_no=str(roll_no) # 100 -> "100"
print("my address is" +" " +address + " "+"and roll_no is" +" "+ roll_no)
print("my address is", address ,"and roll_no is", roll_no)
# my address is pune and roll_no is 100
```

## format () Method

```python
roll_no=100
address="pune"
pin_code=412105
print("my address is: {} and roll_no is: {} and pin code is {}".format(address,roll_no,pin_code))
# my address is pune and roll_no is 100 and pin code is 412105
```

```python
name="rahul"
age=25
print("Name:{}, Age:{}".format(name,age))
# Name:rahul, Age:25
```

## f-string Method

```python
roll_no=100
address="pune"
pin_code=412105
print(f"my address is {address} and roll_no is {roll_no} and pin code is {pin_code}")
# my address is pune and roll_no is 100 and pin code is 412105
```
